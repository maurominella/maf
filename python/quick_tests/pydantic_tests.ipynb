```python
from pydantic import BaseModel, Field

class EchoResponse(BaseModel):
    echo: str

# Example usages of Pydantic model instantiation
my_dict = {"echo": "Dict1", "echo": "1"}
x = EchoResponse(echo="Test")
y = EchoResponse(**my_dict)
z = EchoResponse(**{"echo": "Dict2"})

x
```

```python
class employee(BaseModel):
    id: str = Field(json_schema_extra={"description": "The employee ID"})
    age: int= Field(json_schema_extra={"description": "The employee age"})
    name: str= Field(json_schema_extra={"description": "The employee name", "alias":"first_name"})


e_json = {"id": "ABCD1234", "age": 30, "name": "Mauro"}

emp = employee(**e_json)

emp
```

```python
employee.model_validate(e_json)
```
